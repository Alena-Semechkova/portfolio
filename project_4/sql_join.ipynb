### 1. OUTER JOIN:
Вам доступна база данных, принадлежащая метеоресурсу "Гисметео". Она включает таблицы Countries и Weather. Первая таблица хранит информацию о странах, вторая — информацию об ожидаемой температуре в этих странах в определенные дни.


````python
+----+------------+
| id | name       |
+----+------------+
| 1  | Russia     |
| 2  | Kazakhstan |
| 3  | China      |
| 4  | Ukraine    |
| 5  | Belarus    |
+----+------------+
````

````python
+------------+---------------+------------+
| country_id | weather_state | day        |
+------------+---------------+------------+
| 4          | -20           | 2023-01-12 |
| 1          | 13            | 2023-03-19 |
| 3          | -7            | 2022-10-22 |
| 1          | -7            | 2022-11-14 |
| 5          | 30            | 2023-01-06 |
| 5          | 12            | 2023-04-27 |
| 5          | 6             | 2022-11-20 |
| 3          | 20            | 2023-09-30 |
| 2          | 22            | 2023-07-06 |
| 1          | 13            | 2023-06-01 |
| 3          | 30            | 2023-09-17 |
| 4          | 23            | 2023-09-09 |
| 5          | 7             | 2023-03-01 |
| 2          | 32            | 2023-08-25 |
| 3          | 23            | 2022-09-27 |
| 2          | -29           | 2023-01-05 |
| 3          | 17            | 2023-03-15 |
| 1          | 1             | 2023-07-20 |
| 1          | 0             | 2022-10-08 |
| 4          | 15            | 2023-09-27 |
| 3          | 15            | 2023-07-07 |
| 1          | -12           | 2022-11-12 |
| 2          | -5            | 2023-04-08 |
| 2          | 15            | 2023-09-18 |
| 5          | 13            | 2022-09-28 |
| 3          | 8             | 2023-03-01 |
| 4          | -10           | 2023-03-12 |
| 5          | 20            | 2023-09-01 |
| 5          | 25            | 2023-09-04 |
| 3          | 10            | 2023-08-28 |
+------------+---------------+------------+
````

В зависимости от средней температуры в месяце, метеоресурс характеризует ее определенным типом погоды:

- Cold, если средняя температура меньше или равна 15  
- Hot, если средняя температура больше или равна 25  
- Warm в остальных случаях

**Напишите запрос, который извлекает из предложенной базы данных названия стран, а также определяет тип погоды в каждой стране в сентябре 2023 года. При этом в результирующую таблицу должны быть добавлены названия только тех стран, информация о температуре в которых известна.**

Поле с названием страны должно иметь псевдоним country, поле с типом погоды в стране — weather_type.

#### Решение:

``` sql
SELECT Countries.name AS country, CASE
                           WHEN AVG(Weather.weather_state) <= 15 THEN 'Cold'
                           WHEN AVG(Weather.weather_state) >= 25 THEN 'Hot'
                           ELSE 'Warm'
                           END AS weather_type
FROM Countries RIGHT OUTER JOIN Weather ON Countries.id = Weather.country_id 
WHERE YEAR(Weather.day) = 2023 AND MONTH(Weather.day) = 09  
GROUP BY Countries.name 
```

Первое поле этой таблицы содержит идентификатор пользователя, запросившего сообщение с подтверждением регистрации, второе — дату и время запроса, третье — информацию о том, была ли подтверждена регистрация (confirmed — подтверждено, timeout — истек срок подтверждения).

**Напишите запрос, извлекающий из предложенной базы данных идентификаторы пользователей, которые как минимум дважды запросили сообщение с подтверждением регистрации в течение 24 часов.**

#### Решение:

``` sql
SELECT DISTINCT C1.user_id 
FROM Confirmations AS C1 JOIN Confirmations AS C2 ON C1.user_id = C2.user_id
WHERE C1.time_stamp != C2.time_stamp AND C1.time_stamp < C2.time_stamp AND TIMESTAMPDIFF(SECOND, C1.time_stamp, C2.time_stamp) <= 86400
```

Первое поле этой таблицы содержит идентификатор сессии, второе — длительность сессии в секундах. Например, первая запись таблицы Sessions говорит о том, что некоторый пользователь провел в приложении 264 секунды.
**Напишите запрос, который определяет количество сессий продолжительностью:**  
- [0-5) — от 0 (включительно) до 5 минут (не включительно)
- [5-10) — от 5 (включительно) до 10 минут (не включительно)
- [10-15) — от 10 (включительно) до 15 минут (не включительно)
- 15 or more — от 15 минут (включительно) и более

Поле с продолжительностью сессии должно иметь псевдоним session_duration, поле с количеством сессий — total.

``` sql
SELECT '[0-5)' AS session_duration, COUNT(*) AS total
FROM Sessions
WHERE duration < 300

UNION

SELECT '[5-10)' AS session_duration, COUNT(*) AS total
FROM Sessions
WHERE  duration >= 300 AND duration < 600

UNION 

SELECT '[10-15)' AS session_duration, COUNT(*) AS total
FROM Sessions
WHERE  duration >= 600 AND duration < 900

UNION 

SELECT '15 or more' AS session_duration, COUNT(*) AS total
FROM Sessions
WHERE  duration >= 900
```
